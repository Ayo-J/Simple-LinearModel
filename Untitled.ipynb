```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
X= df[['year']]
y = df['per capita income (US$)']
```

```python
%matplotlib inline
plt.scatter(X,y,color='red')
```

```python
model = LinearRegression()

model.fit(df[["year"]], df["per capita income (US$)"])
```

```python
# Predict per capita income for the year 2020
predicted_income_2020 = model.predict([[2020]])

print(f"Predicted per capita income for the year 2020: ${predicted_income_2020[0]:.2f}")
```

```python
plt.scatter(df["year"], df["per capita income (US$)"], color='blue', label='Actual data')

# Adding the prediction for 2020 to the plot
plt.scatter([2020], predicted_income_2020, color='green', marker='x', s=100, label='Predicted 2020')
```

```python
new = model.predict([[2002]])
```

```python
# Scatter plot for the prediction
plt.scatter(2002, new, color='green', marker='x', s=100, label='Predicted Income for 2002')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Year vs. Per Capita Income')
plt.legend()
```
